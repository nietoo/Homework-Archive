```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC


linSVC = Pipeline(steps=[
    ('scaling', StandardScaler()),
    ('svc', SVC(random_state=413, tol=1e-2, kernel='linear', max_iter=3000))
])

parameters = np.logspace(-3, 3, 7)
plt.title('Regularization coefficient search grid')
searcher = GridSearchCV(linSVC, [{"svc__C": parameters}], scoring="roc_auc", cv=5, n_jobs=-1)
searcher.fit(X_train, y_train)
plt.xlabel('C coefficient')
plt.ylabel('AUC-ROC score')


best_c_svc = searcher.best_params_["svc__C"]

print('Best C: ', best_c_svc)

plt.xscale('log')
plt.plot(parameters, searcher.cv_results_["mean_test_score"])
plt.show()
```

Перейдём к оценке качества выдаваемых алгоритмами вероятностей. Начнём с калибровочных кривых. 

Допустим, алгоритм возвращает некоторые числа от нуля до единицы. Хорошо ли они оценивают вероятность? Для этого разобьем отрезок $[0, 1]$ на несколько маленьких отрезков одинаковой длины. Рассмотрим $i$-й отрезок с границами $[a_i, b_i]$ и предсказания $p_1, p_2, \dots, p_k$, которые попали в него. Пусть им соответствуют истинные ответы $y_1, y_2, \dots, y_k$. Если алгоритм выдает корректные вероятности, то среди этих истинных ответов должно быть примерно $(a_i + b_i) / 2$ единиц. Иными словами, если нарисовать кривую, у которой по оси X отложены центры отрезков, а по оси Y — доли единичных ответов этих в отрезках, то она должна оказаться диагональной. Ниже приведена функция, которая должна рисовать такие кривые. В ней допущено две ошибки — найдите и исправьте их.

```python
def plot_calibration_curve(y_test, preds, label=''):
    bin_middle_points = []
    bin_real_ratios = []
    n_bins = 10
    for i in range(n_bins):
        l = 1.0 / n_bins * i
        r = 1.0 / n_bins * (i + 1)
        bin_middle_points.append((l + r) / 2) 
        bin_real_ratios.append(np.mean(y_test[(preds >= l) & (preds < r)] == 1))
    # чтобы можно было несколько графиков на одном плоте делать
    # plt.figure(figsize=(6,6))
    plt.plot(bin_middle_points, bin_real_ratios,label=label)
    plt.ylim([-0.05, 1.05])
```

# Часть 1. Работа с переменными.

В этой части мы научимся обрабатывать категориальные переменные, так как закодировать их в виде чисел недостаточно (это задаёт некоторый порядок, которого на категориальных переменных может и не быть). Существует два основных способа обработки категориальных значений:
- One-hot-кодирование
- Счётчики (CTR, mean-target кодирование, ...) — каждый категориальный признак заменяется на среднее значение целевой переменной по всем объектам, имеющим одинаковое значение в этом признаке.

Начнём с one-hot-кодирования. Допустим наш категориальный признак $f_j(x)$ принимает значения из множества $C=\{c_1, \dots, c_m\}$. Заменим его на $m$ бинарных признаков $b_1(x), \dots, b_m(x)$, каждый из которых является индикатором одного из возможных категориальных значений:
$$
b_i(x) = [f_j(x) = c_i]
$$

```python
#до начала звонка мы не можем понять, сколько он продлится

X = df.drop(columns=['duration', 'y'])
y = df['y']
y = df['y'].transform(lambda x: 1 if x == 'yes' else 0)



categorical_features = list(X.dtypes[X.dtypes == "object"].index)
numeric_features =  X.select_dtypes([np.number]).columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from time import time

start = time()

parameters = np.logspace(-5, 3, 9)
best_c_lr = 1e-4
searching = True

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=777, stratify = y)

ord_column_transformer = ColumnTransformer([
    ('encoding', OrdinalEncoder(), categorical_features),
    ('scaling', StandardScaler(), numeric_features)
])


ord_lr_pipe = Pipeline(steps=[
    ('encoding_scaling', ord_column_transformer),
    ('lr', LogisticRegression(penalty='l2', random_state=413, solver='liblinear'))
])

if searching:
    searcher = GridSearchCV(ord_lr_pipe, [{"lr__C": parameters}],
                            scoring="roc_auc", cv=5, n_jobs=-1)
    searcher.fit(X_train, y_train)

    plt.title('Regularization strength search grid')
    plt.xscale('log')
    plt.plot(parameters, searcher.cv_results_["mean_test_score"])
    plt.xlabel('C coefficient')
    plt.ylabel('AUC-ROC score')

    best_c_lr = searcher.best_params_["lr__C"]

ord_lr_pipe.set_params(lr__C = best_c_lr)

ord_lr_pipe.fit(X_train, y_train)

y_pred = ord_lr_pipe.predict_proba(X_test)[:,1]

precision, recall, _ = precision_recall_curve(y_test, y_pred)
print('time elapsed:', time() - start, 'sec')
print("AUC-PR = %.5f" % average_precision_score(y_test, y_pred))
print('used', best_c_lr, 'as C')
```

Эту проблему решает другой способ кодирования категориальных признаков — счётчики. Основная идея в том, что нам важны не сами категории, а значения целевой переменной, которые имеют объекты этой категории. Каждый категориальный признак мы заменим средним значением целевой переменной по всем объектам этой же категории:
$$
g_j(x, X) = \frac{\sum_{i=1}^{l} [f_j(x) = f_j(x_i)][y_i = +1]}{\sum_{i=1}^{l} [f_j(x) = f_j(x_i)]}
$$

__Бонус (1 балл)__: Здесь и далее реализуйте вычисление счетчиков с помощью трансформера (наследуйтесь от классов `BaseEstimator, TransformerMixin` из `sklearn.base`). Обратите внимание, что все вычисления должны быть векторизованными, трансформер не должен модифицировать передаваемую ему выборку inplace, а все необходимые статистики нужно считать только по обучающей выборке в методе `fit`. Ваш трансформер должен принимать при инициализации список из категориальных признаков и изменять только их. 


```python
from sklearn.base import BaseEstimator, TransformerMixin
from numpy import linalg


class CTR_Transformer(BaseEstimator, TransformerMixin):
    def __init__(self, features, noise_deviation = 0, smoothing = 0):
        self.features = features
        self.noise_deviation = noise_deviation
        self.smoothing = smoothing
    
    def fit(self, X, y):     
        self.feature_dicts = {}
        self.global_mean = np.mean(y)
        for feature in self.features:
            self.feature_dicts[feature] = ((X.join(y).groupby(feature)['y'].sum() + self.smoothing * self.global_mean) / 
                                            (X.join(y).groupby(feature)['y'].count() + self.smoothing))
        return self
        
    
    def transform(self, X, y = None):
        out = X.copy()
        for feature in self.features:
            out[feature] = out[feature].map(self.feature_dicts[feature])
            if self.noise_deviation:
                np.random.seed(413)
                out[feature] += np.random.normal(0, self.noise_deviation, out[feature].shape[0])
            out[feature] = out[feature].fillna(out[feature].mean()) # In case X_test contains values that were not present in X_train
        return out
```

```python
start = time()


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=777, stratify = y)

ctr_column_transformer = ColumnTransformer([
    ('encoding', CTR_Transformer(categorical_features), categorical_features),
    ('scaling', StandardScaler(), numeric_features)
])

ctr_lr_pipe = Pipeline(steps=[
    ('encoding_scaling', ctr_column_transformer),
    ('lr', LogisticRegression(penalty='l2', random_state=413, solver='liblinear'))
])

if searching:
    searcher = GridSearchCV(ctr_lr_pipe, [{"lr__C": parameters}],
                            scoring="roc_auc", cv=5, n_jobs=-1)
    searcher.fit(X_train, y_train)

    plt.title('Regularization strength search grid')
    plt.xscale('log')
    plt.plot(parameters, searcher.cv_results_["mean_test_score"])
    plt.xlabel('C coefficient')
    plt.ylabel('AUC-ROC score')

    best_c_lr = searcher.best_params_["lr__C"]

ctr_lr_pipe.set_params(lr__C = best_c_lr)

ctr_lr_pipe.fit(X_train, y_train)

y_pred = ctr_lr_pipe.predict_proba(X_test)[:,1]

print('time elapsed:', time() - start, 'sec')
print("AUC-PR = %.5f" % average_precision_score(y_test, y_pred))
print('used', best_c_lr, 'as C')
```

А теперь ответьте на следующий вопрос: что будет, если некоторая категория встречается в выборке всего несколько раз? По этой причине производится сглаживание счётчиков. Например, на практике хорошие результаты показывает использование сглаживания средним по всей выборке:
$$
g_j(x, X) = \frac{\sum_{i=1}^{\ell} [f_j(x) = f_j(x_i)][y_i = +1] + C \times global\_mean}{\sum_{i=1}^{\ell} [f_j(x) = f_j(x_i)] + C}
$$
где $global\_mean$ — доля объектов положительного класса в выборке, $C$ — параметр, определяющий степень сглаживания (например, можно использовать 10 или подобрать для каждого признака свой). Основная идея в том, что мы "разбавляем" среднее значение по некоторой категории глобальным средним значением. И тем меньше, чем большее количество объектов этой категории встречается в выборке. 

Однако для сглаживания вместо среднего значения целевой переменной можно использовать любое другое значение от 0 до 1 (этот параметр иногда называют $prior$). Можно сделать несколько признаков с разными значениями параметра. На практике в задачах бинарной классификации полезными бывают даже отрицательные значения!

### Методы фильтрации


Давайте отбирать признаки умнее, а именно через подсчёт некоторой функции для каждого признака. На основании значений этой функции будем оставлять наиболее важные признаки. Методы этого семейства называют фильтрующими или  filter methods. 

В качестве такой функции будем считать t-статистику:

$$t(j) = \frac{|\mu_+ - \mu_-|}{\sqrt{\frac{n_+ \sigma^2_+ + n_- \sigma^2_-}{n_+ + n_-}}},$$

где $\mu$, $\sigma$, $n$ соответственно среднее, стандартное отклонение и количество объектов каждого из классов.

__Задание 9 (0.5 балла).__ Оставьте 40 признаков с наибольшим значением $t$ и замерьте качество. Не забудьте замерить скорость отбора признаков в этом случае.



```python
from math import sqrt

X = df.drop(columns=['duration', 'y'])
y = (df.y == 'yes')

numeric_features =  X.select_dtypes([np.number]).columns
categorical_features = list(X.dtypes[X.dtypes == "object"].index)

X = pd.concat([X, pd.get_dummies(X[categorical_features])], axis=1).drop(columns=categorical_features)

temp = X.join(y)
t = {}

start = time()

for feature in X.columns:
    mean_pos = temp[temp['y'] == 1][feature].mean()
    mean_neg = temp[temp['y'] == 0][feature].mean()
    std_pos = temp[temp['y'] == 1][feature].std()
    std_neg = temp[temp['y'] == 0][feature].std()
    count_pos = temp[temp['y'] == 1][feature].count()
    count_neg = temp[temp['y'] == 0][feature].count()
    t[feature] = (abs(mean_pos - mean_neg) / 
                  sqrt(((count_pos * (std_pos ** 2)) + (count_neg * (std_neg ** 2))) / (count_pos + count_neg) ))
    
coefs = pd.DataFrame()
coefs['names'] = X.columns
coefs['values'] = coefs['names'].map(t)
coefs = coefs.sort_values(by=['values'], axis=0, ascending = False)
coefs = coefs.iloc[0:40, :]

X = X[list(coefs['names'])]
print('feature filtering complete, time elapsed:', time() - start, 'sec')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=777, stratify=y)


ctr_column_transformer = ColumnTransformer([
    ('scaling', StandardScaler(), numeric_features)
], remainder='passthrough')

ctr_lr_pipe = Pipeline(steps=[
    ('encoding_scaling', StandardScaler()),
    ('lr', LogisticRegression(penalty='l2', random_state=413, solver='liblinear'))
])

if searching:
    searcher = GridSearchCV(ctr_lr_pipe, [{"lr__C": parameters}],
                            scoring="roc_auc", cv=5, n_jobs=-1)
    searcher.fit(X_train, y_train)

    plt.title('Regularization strength search grid')
    plt.xscale('log')
    plt.plot(parameters, searcher.cv_results_["mean_test_score"])
    plt.xlabel('C coefficient')
    plt.ylabel('AUC-ROC score')

    best_c_lr = searcher.best_params_["lr__C"]

ctr_lr_pipe.set_params(lr__C = best_c_lr)

ctr_lr_pipe.fit(X_train, y_train)

y_pred = ctr_lr_pipe.predict_proba(X_test)[:,1]

print('time elapsed:', time() - start, 'sec')
print("AUC-PR = %.5f" % average_precision_score(y_test, y_pred))
print('used', best_c_lr, 'as C')
```
